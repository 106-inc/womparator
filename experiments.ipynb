```python
import os

def extract_images(docx_file, output_dir):
    document = docx.Document(docx_file)
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    for i, inline_shape in enumerate(document.inline_shapes):
        image_part = inline_shape.part
        image_filename = f"image_{i}.{image_part.partname.split('.')[-1]}"
        image_path = os.path.join(output_dir, image_filename)

        with open(image_path, 'wb') as f:
            image_part.blob.seek(0)  # Reset to beginning of image data
            f.write(image_part.blob)

extract_images(infile, "extracted_images")
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from os import listdir
from os.path import isfile, join

image_dir = 'extracted_images'
image_files = [f for f in listdir(image_dir) if isfile(join(image_dir, f))]

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))  # Adjust layout as needed

for i, image_file in enumerate(image_files[:4]):  # Limit to 4 images for this example
    img = mpimg.imread(os.path.join(image_dir, image_file))
    row, col = i // 2, i % 2
    axes[row, col].imshow(img)
    axes[row, col].set_title(image_file)
    axes[row, col].axis('off')

plt.tight_layout()
plt.show()
```
